## Training and Test Dataset Split 

```python
prices = data['PRICE']
features = data.drop('PRICE', axis=1)

X_train, X_test, y_train, y_test = tts(features, prices, test_size=0.2, random_state=10)
```

## Multivariable Regression

```python
regr = LinearRegression()
regr.fit(X_train, y_train)

print('Intercept', regr.intercept_)
pd.DataFrame(data=regr.coef_, index=X_train.columns, columns=['coef'])
```

```python
print('Training data r-squared:', regr.score(X_train, y_train))
print('Testing data r-squared:', regr.score(X_test, y_test))
```

## p values and evaluating Co-efficients

```python
X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)

results = model.fit()
#results.params
#results.pvalues

pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})
```

## Model Simplification and the BIC


```python
# Original model with log prices and all features
X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)

results = model.fit()

org_coef = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})
print('BIC is:',results.bic)
print('R-squared is:',results.rsquared)
```

```python
X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['INDUS'], axis=1)
model = sm.OLS(y_train, X_incl_const)

results = model.fit()

coef_minus_indus = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})
print('BIC is:',results.bic)
print('R-squared is:',results.rsquared)
```

```python
X_incl_const = sm.add_constant(X_train)
X_incl_const = X_incl_const.drop(['AGE', 'INDUS'], axis=1)
model = sm.OLS(y_train, X_incl_const)

results = model.fit()

reduced_coef = pd.DataFrame({'coef': results.params, 'p-value': round(results.pvalues, 3)})
print('BIC is:',results.bic)
print('R-squared is:',results.rsquared)
```

```python
frames = [org_coef, coef_minus_indus, reduced_coef]
pd.concat(frames, axis=1)
```

## Residuals and Residual Plots

```python
# Modified Model: Transformed (using log prices) and simplified (dropping two features)
prices = np.log(data['PRICE'])
features = data.drop(['PRICE','INDUS','AGE'], axis=1)

X_train, X_test, y_train, y_test = tts(features, prices, test_size=0.2, random_state=10)

# Using statsmodel
X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)
results = model.fit()

#residuals
#residuals = y_train - results.fittedvalues
#results.resid

# Graph of Actual vs Fitted Values
corr = round(y_train.corr(results.fittedvalues), 2)

plt.figure(figsize=[10,5])
plt.plot(y_train, y_train, color='cyan')
plt.scatter(x=y_train, y=results.fittedvalues, alpha=0.6, color='navy')
plt.xlabel('Actual Log Prices $y _i$', fontsize=14)
plt.ylabel('Predicted Log Prices $\hat y _i$', fontsize=14)
plt.title(f'Actual vs Predicted Log Prices $y _i$ vs $\hat y _i$ (Corr {corr} )', fontsize=17)
plt.show()

plt.figure(figsize=[10,5])
plt.plot(np.e**y_train, np.e**y_train, color='cyan')
plt.scatter(x=np.e**y_train, y=np.e**results.fittedvalues, alpha=0.6, color='navy')
plt.xlabel('Actual Prices 000s $y _i$', fontsize=14)
plt.ylabel('Predicted Prices 000s $\hat y _i$', fontsize=14)
plt.title(f'Actual vs Predicted Prices $y _i$ vs $\hat y _i$ (Corr {corr} )', fontsize=17)
plt.show()

#Residuals vs Predicted values
plt.figure(figsize=[10,5])
plt.scatter(x=results.fittedvalues, y=results.resid, alpha=0.6, color='navy')
plt.xlabel('Actual Prices', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted values', fontsize=17)
plt.show()

#Mean Squared Error & R-Squared Error
reduced_log_mse = round(results.mse_resid, 3)
reduced_log_rsquared = round(results.rsquared, 3)
```

```python
resid_mean = round(results.resid.mean(), 3)
resid_skew = round(results.resid.skew(), 3)

sns.distplot(results.resid, color='navy')
plt.title(f'Log price model : Residuals Skew ({resid_skew}) Mean ({resid_mean})')
plt.show()
```

```python
prices = data['PRICE']
features = data.drop(['PRICE'], axis=1)

X_train, X_test, y_train, y_test = tts(features, prices, test_size=0.2, random_state=10)

# Using statsmodel
X_incl_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_incl_const)
results = model.fit()

# Graph of Actual vs Predicted Values
corr = round(y_train.corr(results.fittedvalues), 2)

plt.figure(figsize=[10,5])
plt.plot(y_train, y_train, color='cyan')
plt.scatter(x=y_train, y=results.fittedvalues, alpha=0.6, color='navy')
plt.xlabel('Actual Prices', fontsize=14)
plt.ylabel('Predicted Prices', fontsize=14)
plt.title(f'Actual vs Predicted (Corr {corr} )', fontsize=17)
plt.show()

#Residuals vs Predicted values
plt.figure(figsize=[10,5])
plt.scatter(x=results.fittedvalues, y=results.resid, alpha=0.6, color='navy')
plt.xlabel('Actual Prices', fontsize=14)
plt.ylabel('Residuals', fontsize=14)
plt.title('Residuals vs Fitted values', fontsize=17)
plt.show()

res_mean = round(results.resid.mean(), 3)
res_skew = round(results.resid.skew(), 3)

sns.distplot(results.resid, color='navy')
plt.title(f'Residuals Skew ({res_skew}) Mean ({res_mean})')
plt.show()

#Mean Squared Error & R-Squared Error
full_normal_mse = round(results.mse_resid, 3)
full_normal_rsquared = round(results.rsquared, 3)
```
